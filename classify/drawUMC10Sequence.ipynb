```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Global settings
plt.rcParams.update({
    'font.size': 12,  # Set global font size
    'axes.labelsize': 24,  # Label size
    'axes.titlesize': 24,  # Title size
    'xtick.labelsize': 24,  # X-axis tick size
    'ytick.labelsize': 24,  # Y-axis tick size
    'font.family': 'Arial',  # Change to Arial font
    'text.usetex': False  # Ensure LaTeX is used correctly
})

# Set seaborn style and color palette
sns.set_palette("deep")
sns.set_style("whitegrid", {'axes.grid': False, 'ytick.left': True})

# Define colors
colors = ["#74AED4", "#F7C97E", "#ECA8A9", "#C6CF9D"]  # Blue, Orange, Green, Red

# Data for the two cases
categories = ["2Ã—10$^{5}$", "No Limit"]  # Use LaTeX formatting for superscript
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
values = [
    [0.0542, 0.0079],  # Accuracy
    [0.4842, 0.0094],  # Precision
    [0.0575, 0.0480],  # Recall
    [0.1028, 0.0157]   # F1 Score
]

# Convert to numpy array for easier handling
values = np.array(values)

# Bar width
width = 0.2
x = np.arange(len(categories))  # X-axis positions

# Plotting
plt.figure(figsize=(8, 6))

# Plot each metric as a bar and add value labels
for i, metric in enumerate(metrics):
    bars = plt.bar(x + i * width - width * (len(metrics) / 2), values[i], width, 
                   label=metric, 
                   color=colors[i],
                   alpha=0.5,
                   edgecolor="#555555")
    
    # Add value labels on top of bars
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom',
                 fontsize=24, fontweight='bold')  # Use global font size and bold text

# Labels and Title
plt.xticks(x, categories)
# plt.ylabel("Score")
# plt.legend()

# Remove grid and unnecessary spines
plt.grid(False)
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Save the plot as a SVG file
plt.savefig("UMC10Sequence.svg", format="svg", bbox_inches="tight")

# Show the plot
plt.show()
```
